```python
import pandas as pd
import numpy as np
```

## This notebook processes input hobert file and outputs the validation dataframe

```python
df = pd.read_csv('markerhobertetal.csv', sep='\t')
/print df.shape
display(df.head())
```

```python
geneColumns = df.columns[4:]
genes = df[geneColumns].values.reshape(-1)
genes = [i for i in genes if str(i) != 'nan']
genes = np.unique(genes)
print(f"Found {len(genes)} unique genes")
```

```python
neurons = []
prevNeuron = df['Neuron class'].values[0]
neurons.append(prevNeuron)
for n in df['Neuron class'].values[1:]:
    if str(n) != 'nan':
        prevNeuron = n 
    neurons.append(prevNeuron)
df['class'] = neurons
df[['Neuron class', 'class']]
```

```python
df = df[np.concatenate([['class'], geneColumns])]
df.head()
```

## Rename gene column names

```python
geneColumns = [f"gene{i}" for i in range(len(geneColumns))]
df.columns = np.concatenate([['class'], geneColumns])
df.head()
```

```python
result = pd.DataFrame(columns=np.concatenate([['class'], genes]))
result.head()
```

```python
for c in df['class'].unique():
    classGenes = df[df['class'] == c][geneColumns].values.reshape(-1)
    classGenes = [i for i in classGenes if str(i) != 'nan']
    classGenes = np.unique(classGenes)
    classGenes = np.isin(genes, classGenes).astype(int)
    classGenes = np.concatenate([[c], classGenes])
    result.loc[result.shape[0]] = classGenes
    
#     genes[np.isin(classGenes, geneColumns)]
#     print(f'class {c}, genes : {classGenes}')
```

```python
result[genes] = result[genes].astype(int)
```

```python
result.head()
```

```python
result.to_pickle('data/horbert.pkl')
```
